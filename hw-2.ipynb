```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = np.pi
n = 2*pi

x = np.arange(0,n,n/1000)   #makes array from 0 to 2*pi with 1000 elements
y = (5.5)* np.cos(2*x)+5.5  #a
plt.plot(x,y)               #creates plot
y = (0.02)*np.exp(x)        #b
plt.plot(x,y)
y = (0.25) * (np.power(x,2)) + (0.1)*np.sin(10*x) #c
plt.plot(x,y)

plt.xlabel("Time in ASTR/EART 119")
plt.ylabel("Measures of Awesomeness")
plt.xlim([0,n])             #sets range
plt.ylim([-1,10])
plt.show()
```
